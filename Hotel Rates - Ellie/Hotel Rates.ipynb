```python
import pandas as pd
compset_names = soup.findAll('span', class_ = "item-link name__copytext")
compset_prices = soup.findAll('strong', class_ = "item__best-price price_min")
l = []
for i in compset_names:
    l.append(i.text)
r = []
for i in compset_prices:
    r.append(i.text)

df = pd.DataFrame({'Hotels':l,'Price':r})
df
```

```python
l = []
for i in df.Price:
    x = float(i.replace('$','').replace(',',''))
    l.append(x)
    
df['Price'] = l
df
```

```python
benchmark = df['Price'].mean()
benchmark
```

```python
print("The Average Daily Rate your hotel can charge is HK$" + str(round(benchmark)))
```
