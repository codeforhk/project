```python
from selenium import webdriver
driver = webdriver.Chrome('C:/Users/ellie/Documents/chromedriver/chromedriver.exe')
driver.get('https://www.trivago.hk/en?aDateRange%5Barr%5D=2020-01-15&aDateRange%5Bdep%5D=2020-01-16&aPriceRange%5Bfrom%5D=0&aPriceRange%5Bto%5D=0&iRoomType=7&aRooms%5B0%5D%5Badults%5D=2&cpt2=2434894%2F500%2C1322%2F105&iViewType=0&bIsSeoPage=0&sortingId=6&slideoutsPageItemId=&iGeoDistanceLimit=1592&address=&addressGeoCode=&ra=&sLanguageLocale=UK&iOffset=0')
```

```python
driver.page_source
from bs4 import BeautifulSoup as bs

source = driver.page_source
soup = bs(source, 'lxml')
soup 
```

```python
import pandas as pd
compset_names = soup.findAll('span', class_ = "item-link name__copytext")
compset_prices = soup.findAll('strong', class_ = "item__best-price price_min")
l = []
for i in compset_names:
    l.append(i.text)
r = []
for i in compset_prices:
    r.append(i.text)

df = pd.DataFrame({'Hotels':l,'Price':r})
df
```
