```python
from selenium import webdriver #to get contents from interactive websites
```

```python
driver = webdriver.Chrome()
driver.get('https://www.airbnb.com/s/Hong-Kong/homes?refinement_paths%5B%5D=%2Fhomes&place_id=ChIJByjqov3-AzQR2pT0dDW0bUg&search_type=filter_change&checkin=2019-12-25&checkout=2019-12-28&adults=1')
r = driver.page_source
```

```python
from bs4 import BeautifulSoup as bs
import pandas as pd
soup = bs(r, 'lxml') #to gather information of webpage

#Identify property name
property_name = soup.find_all(class_ = '_1ebt2xej')
property_name
```

```python
#Identify nightly price
rating = soup.find_all(class_ = '_ky9opu0')
rating
```

```python
#Identify total price
total_price = soup.find_all(class_ = '_1llb8an')
total_price
```

```python
#put ratings information onto a dataframe "j"
j = []
for x in rating:
    j.append(x.text)
pd.DataFrame(j)
```

```python
#put property information onto a dataframe "l"
l = []
for i in property_name:
    l.append(i.text)
pd.DataFrame(l)
```
